```
from fastai.vision import *
from fastai.metrics import error_rate
```

```
data = ImageDataBunch.from_folder(Path("/content/drive/My Drive/Photos"),valid_pct=0.1, ds_tfms=get_transforms(), size=224,num_workers=2, bs=12)
data.show_batch(rows=5)
```

```
data.classes
```

```
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```
learn.fit_one_cycle(4)
```

```
img=open_image(Path('/content/_102815470_racoon4.jpg'))
img
```

```
pred_class,pred,outputs=learn.predict(img)
pred_class.obj
```
